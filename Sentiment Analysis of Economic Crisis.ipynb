```python
data = pd.read_csv("C:/Users/sharn/Desktop/DM/ecocrisis.csv")
data.columns
data
```

```python
nltk.download('vader_lexicon')
sentiments = SentimentIntensityAnalyzer()
data["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in data["Transcript"]]
data["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in data["Transcript"]]
data["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in data["Transcript"]]
data['Compound'] = [sentiments.polarity_scores(i)["compound"] for i in data["Transcript"]]
score = data["Compound"].values
sentiment = []
for i in score:
    if i >= 0.05 :
        sentiment.append('Positive')
    elif i <= -0.05 :
        sentiment.append('Negative')
    else:
        sentiment.append('Neutral')
data["Sentiment"] = sentiment
data
```

```python
count1=0
count2=0
count3=0
for i in score:
    
    if i>=0.05:
        count1+=1
    elif i <= -0.05:
        count2+=1
    else:
        count3+=1

positive_per=count1/100*100
negative_per=count2/100*100
neutral_per=count3/100*100
print("Positive percentage:",positive_per,"%")
print("Negative percentage:",negative_per,"%")
print("Neutral percentage:",neutral_per,"%")
```
